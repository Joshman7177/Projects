```python
#Create a Matrix/Determinant calculator in Python from scratch using Object Oriented Programming


class Matrix:
    def __init__(self, num, nrow):
        self.list = num
        self.rows = nrow
      
        
        for i in range(self.rows):
            ncol = int(len(num) / self.rows)
            if len(self.list) % self.rows == 0:
                self.mat = [num[(i * ncol):(i * ncol + ncol)] for i in range(nrow)]
                self.dim = [nrow, ncol]                
            else:
                raise ValueError('Invalid Dimensions')
    
    def __str__(self):
        return str(self.mat)
    
    def __repr__(self):
        return str(self.mat)
                  
    def is_sqr(self):
        return self.dim[0] == self.dim[1]
    
    def __add__(self, other):
        if self.dim == other.dim:
            result = []
            for i in range(self.dim[0]):
                for j in range(self.dim[1]):
                    result.append(self.mat[i][j] + other.mat[i][j])
            return Matrix(result, self.dim[0])
        else:
            raise ValueError('Matrix dimensions dont conform')
            
    def __sub__(self, other):
        if self.dim == other.dim:
            result = []
            for i in range(self.dim[0]):
                for j in range(self.dim[1]):
                    result.append(self.mat[i][j] - other.mat[i][j])
            return Matrix(result, self.dim[0])
        else:
            raise ValueError('Matrix dimensions dont conform')
            
    def exrow(self, i):
        if i >= self.dim[0]:
            raise ValueError('Cannot extract row')
        else:
            return self.mat[i]
        
    def excol(self, i):
        if i >= self.dim[1]:
            raise ValueError('Cannot extract column')
        else:
            return [self.mat[j][i] for j in range(self.dim[0])]
        
    def __pow__(self, other):
        if self.dim[1] != other.dim[0]:
            raise ValueError('Uneven dimensions: Cannot conduct matrix multiplication')
        else:
            result2 = []
            for i in range(self.dim[0]):
                row = self.exrow(i)
                for j in range(other.dim[1]):
                    col = other.excol(j)
                    aux = 0
                    for k in range(len(row)):
                        aux += row[k] * col[k]
                    result2.append(aux)
                        
            return Matrix(result2, self.dim[0])
  
    def subm(self, row, col):
        if self.is_sqr():
            #Do the thing
            result3 = []
            for i in range(self.dim[0]):
                for j in range(self.dim[1]):
                    if i != row and j != col:
                        result3.append(self.mat[i][j])
            return Matrix(result3, self.dim[0]-1)                
        else:
            raise ValueError('Cannot take Submatrix')
            
    def det(self):
        if self.is_sqr():
            if self.dim[0] == 1:
                return self.mat[0][0]
            else:
                res = 0
                for i in range(self.dim[0]):
                    res += self.mat[i][0] * ((-1) ** i) * self.subm(i, 0).det()
                return res
   
    def t(self):
        res = []
        for i in range(self.dim[1]):
            res += self.excol(i)
        return Matrix(res, self.dim[1])
    
    def cofm(self):
        if self.is_sqr():
            res = list()
            for i in range(self.dim[1]):
                for j in range(self.dim[0]):
                    res += self.subm(int(i),int(j)).det()
            return Matrix(res, self.dim[1])
        else:
            raise ValueError('Cannot compute cofactor matrix')
            
    def inv(self):
        res = []
        if self.is_sqr():
            cofact = self.cofm()
            for i in cofact.t():
                res += i % self.det()
            return Matrix(res, self.dim[1])
        else:
            raise ValueError('Cannot compute inverse Matrix')
    

m1 = Matrix([1,2,3,1], 2)
m2 = Matrix([2,3,5,1,3,5], 3)
```
